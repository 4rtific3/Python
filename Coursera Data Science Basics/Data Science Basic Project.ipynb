```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patch
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("csv files/Banknote-authentication-dataset-.csv")
V1 = data["V1"]
V2 = data["V2"]
data
```

```python
data.describe()
```

```python
array = np.array(data)
array
```

```python
mean = array.mean(0)
mean
```

```python
stdev = array.std(0)
stdev
```

```python
plt.scatter(V1, V2)
plt.scatter(mean[0], mean[1])
plt.show()
```

```python
array_min = array.min(0)
array_max = array.max(0)
normed = (array - array_min) / (array_max - array_min)
print(normed)
```

```python
plt.scatter(normed[:,0], normed[:,1])
plt.show()
```

```python
array_min = array.min(0)
array_max = array.max(0)
normed = (4 * (array - array_min) / (array_max - array_min)) - 2

plt.scatter(normed[:,0], normed[:,1])
plt.show()
```

```python
V1_V2 = np.column_stack((V1,V2))
km = KMeans(n_clusters=2).fit(V1_V2)
clusters = km.cluster_centers_
print(clusters)
```

```python
plt.scatter(V1, V2)
plt.scatter(clusters[:,0], clusters[:,1])
plt.show()
```

Not standardising the data resulted in a skewed weighting of the standard deviation and hence closer K-Means values

```python
V1_V2 = np.column_stack((normed[:,0],normed[:,1]))
km = KMeans(n_clusters=2).fit(V1_V2)
clusters = km.cluster_centers_
plt.scatter(normed[:,0],normed[:,1])
plt.scatter(clusters[:,0], clusters[:,1])
plt.show()
```
